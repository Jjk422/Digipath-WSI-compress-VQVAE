```python
from PIL import Image
import os
```

```python
all_patch_file_paths = os.listdir("JPEG_VQVAE_histopathology_compression/Datasets/Camelyon 2016 subset/thresholded_patches/Camelyon_2016/512/Normal")
patches_per_image = {i.split(".")[0]: [] for i in all_patch_file_paths}

for key, value in patches_per_image.items():
    patches_per_image[key] = [i for i in all_patch_file_paths if i.split(".")[0] == key]

# print(patches_per_image["normal_001"])
print(patches_per_image["normal_001"])
```

```python
def create_canvas_from_filename(filename, background_colour="black"):
    x_max_patches, y_max_patches = [int(i) for i in eval(patches_per_image[filename][0].split(".")[-2].split("_")[-3])]
    
    canvas = Image.new("RGB", (536*x_max_patches, 536*y_max_patches), color=background_colour)

    # canvas.paste(Image.new("RGB", (512, 512), color="white"), (512, 512))

    for filename in patches_per_image[filename]:        
        x_location, y_location = [int(i) for i in filename.split(".")[-2].split("_")[-2:]]

        canvas.paste(Image.open(f"JPEG_VQVAE_histopathology_compression/Datasets/Camelyon 2016 subset/thresholded_patches/Camelyon_2016/512/Normal/{filename}"), (536 * x_location, 536 * y_location))
        
    return canvas
```
